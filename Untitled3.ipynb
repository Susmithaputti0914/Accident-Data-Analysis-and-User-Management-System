```python
import pandas as pd
import numpy as np

# Create a messy dictionary
data = {
    'id': [101, 102, 103, 103, 104, 105, np.nan, 107],  # Contains a duplicate (103) and a missing ID
    'name': ['  Alice ', 'Bob', 'Charlie', 'Charlie', 'David', np.nan, 'Frank', 'Grace'], # Whitespaces and missing name
    'joining_date': ['2023/01/15', '12-05-2022', '2021.08.30', '2021.08.30', '2024/02/10', '05/12/2023', '2022-11-22', '2023-04-01'], # Messy dates
    'salary': [50000, 60000, np.nan, np.nan, 80000, 55000, 62000, 70000] # Missing values
}

# Convert to DataFrame and save as CSV
df_messy = pd.DataFrame(data)
df_messy.to_csv('raw_data.csv', index=False)
print("🎯 'raw_data.csv' has been generated in your Colab files!")
```

```python
import pandas as pd

def clean_dataset(file_path, output_path):
    print("⏳ Loading dataset...")
    df = pd.read_csv(file_path)

    initial_rows = df.shape[0]

    # 1. Remove Duplicate Entries
    df = df.drop_duplicates()
    duplicate_count = initial_rows - df.shape[0]
    print(f"✅ Removed {duplicate_count} duplicate rows.")

    # 2. Handle Missing Values
    df = df.dropna(subset=['id']) # Drop rows where critical 'id' is missing

    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())

    text_cols = df.select_dtypes(include=['object']).columns
    for col in text_cols:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna('Unknown')
    print("✅ Missing values handled.")

    # 3. Fix Inconsistent Formats (Dates & Text)
    date_keywords = ['date', 'timestamp', 'created_at', 'joining']
    for col in df.columns:
        if any(keyword in col.lower() for keyword in date_keywords):
            df[col] = pd.to_datetime(df[col], errors='coerce').dt.strftime('%Y-%m-%d')
            df[col] = df[col].fillna('Unknown')

    for col in text_cols:
        df[col] = df[col].astype(str).str.strip()
    print("✅ Formats standardized to YYYY-MM-DD.")

    # Save the clean data
    df.to_csv(output_path, index=False)
    print(f"🎉 Cleaned dataset saved as: {output_path}")

# Run the cleaning pipeline
clean_dataset('raw_data.csv', 'cleaned_output.csv')
```
