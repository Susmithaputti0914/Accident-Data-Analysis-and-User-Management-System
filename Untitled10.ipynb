```python
import pandas as pd
import io

# 1. Create a messy sample dataset with duplicates and missing values
messy_data = """EmployeeID,Name,Department,Salary,JoinDate
101,Ananya,Sales,50000,2023-01-15
102,Rahul,Marketing,,2023-03-20
103,Sneha,HR,45000,2023-02-10
101,Ananya,Sales,50000,2023-01-15
104,Vikram,,60000,2023-05-12
105,Pooja,Marketing,55000,
103,Sneha,HR,45000,2023-02-10"""

print("--- ORIGINAL MESSY DATA ---")
df = pd.read_csv(io.StringIO(messy_data))
print(df)
print("\n" + "="*40 + "\n")

# 2. AUTOMATION WORKFLOW
# Step A: Remove duplicate rows
df.drop_duplicates(inplace=True)

# Step B: Handle missing values (Fill missing Department with 'Unknown', Salary with average)
df['Department'].fillna('Unknown', inplace=True)
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
df['JoinDate'].fillna('Not Provided', inplace=True)

print("--- AUTOMATED CLEAN DATA ---")
print(df)
print("\n" + "="*40 + "\n")

# 3. GENERATE AUTOMATED REPORT SUMMARY
print("--- AUTOMATED REPORT SUMMARY ---")
summary_report = df.groupby('Department').agg(
    Total_Employees=('EmployeeID', 'count'),
    Average_Salary=('Salary', 'mean')
).reset_index()

print(summary_report)
```
