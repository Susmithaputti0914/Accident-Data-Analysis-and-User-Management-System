```python
import pandas as pd
import plotly.express as px

# 1. Generate a mock business dataset
data = {
    'Quarter': ['Q1', 'Q1', 'Q2', 'Q2', 'Q3', 'Q3', 'Q4', 'Q4'],
    'Department': ['Sales', 'Marketing', 'Sales', 'Marketing', 'Sales', 'Marketing', 'Sales', 'Marketing'],
    'Revenue_USD': [45000, 12000, 54000, 15000, 49000, 18000, 62000, 22000],
    'Expenses_USD': [30000, 14000, 32000, 13000, 31000, 15000, 35000, 18000]
}

df = pd.DataFrame(data)
df['Profit_USD'] = df['Revenue_USD'] - df['Expenses_USD']

print("🎯 Dataset created! Generating dashboard charts...\n")

# Chart A: Interactive Revenue Breakdown by Department
fig_bar = px.bar(
    df,
    x='Quarter',
    y='Revenue_USD',
    color='Department',
    barmode='group',
    title='Quarterly Revenue Growth by Department',
    labels={'Revenue_USD': 'Revenue ($)'},
    template='plotly_dark'
)
fig_bar.show()

# Chart B: Overall Net Profit Margin Trends
# Note: Added numeric_only=True to avoid future warnings during aggregation
df_total = df.groupby('Quarter').sum(numeric_only=True).reset_index()

fig_line = px.line(
    df_total,
    x='Quarter',
    y='Profit_USD',
    markers=True,  # Fixed: Changed from marker='o' to markers=True
    title='Overall Company Profit Trend',
    labels={'Profit_USD': 'Net Profit ($)'},
    template='plotly_dark'
)

# Optional: Customizing the marker circles further via update_traces
fig_line.update_traces(
    line_color='limegreen',
    line_width=3,
    marker=dict(size=10, symbol='circle')
)

fig_line.show()
```
