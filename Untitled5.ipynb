```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_rows = 15

web_data = {
    'session_id': [f"SESS_{100 + i}" for i in range(n_rows)],
    'page_views': [5, 1, 12, 2, np.nan, 8, 1, 15, 3, np.nan, 7, 2, 9, 1, 4],
    'session_duration_secs': [320, 15, 1200, 45, np.nan, 540, 8, 1450, 90, 110, 410, 55, 720, np.nan, 180],
    'stage_reached': ['Purchase', 'Homepage', 'Purchase', 'Cart', 'Homepage',
                      'Product_Page', 'Homepage', 'Purchase', 'Cart', 'Product_Page',
                      'Product_Page', 'Homepage', 'Cart', 'Homepage', 'Product_Page']
}

df_web = pd.DataFrame(web_data)
df_web.to_csv('web_traffic_data.csv', index=False)
print("🎯 'web_traffic_data.csv' has been successfully generated!")
```

```python
import pandas as pd

df = pd.read_csv('web_traffic_data.csv')

df['page_views'] = df['page_views'].fillna(df['page_views'].median()).astype(int)
df['session_duration_secs'] = df['session_duration_secs'].fillna(df['session_duration_secs'].median())

print("✅ Data cleaning complete. Gaps in session data filled dynamically.\n")

avg_views = df['page_views'].mean()
avg_duration = df['session_duration_secs'].mean()
bounce_rate = (len(df[df['page_views'] == 1]) / len(df)) * 100

print(f"📊 --- WEB TRAFFIC ANALYSIS REPORT ---")
print(f"• Total Sessions Evaluated: {len(df)}")
print(f"• Average Page Views per Session: {avg_views:.1f}")
print(f"• Average Session Duration: {avg_duration:.1f} seconds")
print(f"• Estimated Bounce Rate (1 Page View Only): {bounce_rate:.2f}%\n")


print("🔍 User Journey Funnel Stage Volume:")
funnel_stages = df['stage_reached'].value_counts()
print(funnel_stages)

print("\n💡 Insights & Recommendations:")
print("• Drop-off Point: A major segment of users stall at the Homepage and Product Page stages.")
print("• Action Item: Simplify the product page checkout layout and streamline cart navigation to accelerate conversions.")

df.to_csv('cleaned_web_analytics.csv', index=False)
```
