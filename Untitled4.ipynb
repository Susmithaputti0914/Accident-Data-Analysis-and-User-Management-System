```python
import pandas as pd
import numpy as np


churn_data = {
    'customer_id': [1, 2, 3, 4, 5, 6, 7, 7, 9, 10],
    'tenure_months': [12, 2, np.nan, 24, 1, 36, 5, 5, 18, np.nan],
    'monthly_charges': [29.99, 79.99, 59.99, np.nan, 99.99, 45.00, 85.50, 85.50, np.nan, 65.00],
    'total_spent': [359.88, 159.98, 179.97, 2160.00, 99.99, 1620.00, 427.50, 427.50, np.nan, np.nan],
    'contract_type': ['Month-to-month', 'Month-to-month', 'One year', 'Two year', 'Month-to-month', 'Two year', np.nan, np.nan, 'One year', 'Month-to-month'],
    'churn': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes']
}

df_churn = pd.DataFrame(churn_data)
df_churn.to_csv('churn_data.csv', index=False)
print("🎯 'churn_data.csv' has been generated in your Colab files!")
```

```python
import pandas as pd

df = pd.read_csv('churn_data.csv')
df = df.drop_duplicates()
df['tenure_months'] = df['tenure_months'].fillna(df['tenure_months'].median())
df['monthly_charges'] = df['monthly_charges'].fillna(df['monthly_charges'].median())
df['total_spent'] = df['total_spent'].fillna(df['tenure_months'] * df['monthly_charges'])
df['contract_type'] = df['contract_type'].fillna('Month-to-month')

total_customers = len(df)
churned_customers = len(df[df['churn'] == 'Yes'])
churn_rate = (churned_customers / total_customers) * 100

print(f"📊 --- CHURN REPORT ---")
print(f"Total Unique Customers: {total_customers}")
print(f"Churned Customers: {churned_customers}")
print(f"Overall Churn Rate: {churn_rate:.2f}%")

avg_charges = df.groupby('churn')['monthly_charges'].mean()
print(f"• Avg Monthly Bill for Churned Users: ${avg_charges['Yes']:.2f}")
print(f"• Avg Monthly Bill for Retained Users: ${avg_charges['No']:.2f}")
```
