```python
import pandas as pd

# Create a mock geospatial dataset representing potential expansion regions
geo_data = {
    'region_id': ['REG_01', 'REG_02', 'REG_03', 'REG_04', 'REG_05'],
    'region_name': ['Downtown Hub', 'North Suburbs', 'East Commercial Zone', 'South Industrial Park', 'West Residential'],
    'latitude': [40.7128, 40.8075, 40.7128, 40.6500, 40.7484],
    'longitude': [-74.0060, -73.9626, -73.9000, -74.0000, -74.1000],
    'estimated_demand_score': [85, 45, 90, 30, 75],       # Scale out of 100
    'current_competitor_count': [8, 1, 2, 0, 6]            # Low count + high demand = Optimal
}

df_geo = pd.DataFrame(geo_data)
df_geo.to_csv('expansion_data.csv', index=False)
print("🎯 'expansion_data.csv' has been successfully generated!")
```

```python
import pandas as pd
import folium

# 1. Load the data
df = pd.read_csv('expansion_data.csv')

# 2. Identify Optimal Expansion Zones (High Demand > 70 AND Low Competition < 3)
df['is_optimal_zone'] = (df['estimated_demand_score'] > 70) & (df['current_competitor_count'] < 3)

print("📊 --- GEOSPATIAL EXPANSION ANALYSIS ---")
print(df[['region_name', 'estimated_demand_score', 'current_competitor_count', 'is_optimal_zone']])

# 3. Build the Interactive Map Visualization
map_center = [df['latitude'].mean(), df['longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=11)

for _, row in df.iterrows():
    color = 'green' if row['is_optimal_zone'] else 'red'
    icon_type = 'star' if row['is_optimal_zone'] else 'info-sign'

    popup_text = f"""
    <b>{row['region_name']}</b><br>
    Demand Score: {row['estimated_demand_score']}<br>
    Competitors: {row['current_competitor_count']}<br>
    Target Zone: {'YES' if row['is_optimal_zone'] else 'NO'}
    """

    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=folium.Popup(popup_text, max_width=250),
        icon=folium.Icon(color=color, icon=icon_type)
    ).add_to(m)

# Save the map as an interactive HTML file
m.save('expansion_map.html')
print("\n🎉 Interactive expansion map saved as 'expansion_map.html'!")

# Display map directly in your Kaggle window
m
```
